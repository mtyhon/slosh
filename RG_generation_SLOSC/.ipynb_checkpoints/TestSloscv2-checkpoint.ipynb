```python
import astropy.units as u
import lightkurve as lk
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import sloscillations
from tqdm import tqdm
from astropy.units import cds
from sloscillations import generate_data
from joblib import Parallel, delayed
cds.enable()
```

```python
df = pd.read_csv('Marc_metadata_full_v3.csv')
df
```

```python
def create_data(data):
    name, row = data
    t = np.arange(0, row['T'], row['dt'])
    t = (t / 86400.0) * u.day
    #print(t)
    row['calc_rot'] = True
    if row['Phase'] == 0:
        row['evo_state'] = 'RGB'
    elif row['Phase'] == 1:
        row['evo_state'] = 'RC'
    else:
        row['evo_state'] = 'RGB'

    dat = generate_data.GenerateData(row)


    t, y, data, f, model = dat(t)
    
    np.savetxt(f"Test_data_v3/{int(name)}.ts", np.c_[t.value, y])
   
```

```python
results = Parallel(n_jobs=-1)(delayed(create_data)(i) for i in tqdm(df.iterrows(), total=len(df)))
```
